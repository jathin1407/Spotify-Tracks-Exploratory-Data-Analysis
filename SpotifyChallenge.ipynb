## MOST POPULAR ARTIST

```python
Artist_Songs=track.groupby('id_artists').size()
Artist_Songs
```

```python
## TO filter the tracks with artists who have atleast 10 songs
Artist_Songs_count10=Artist_Songs[Artist_Songs >= 10].index
Filtered_Track=track[track['id_artists'].isin(Artist_Songs_count10)]
Filtered_Track
```

```python
##Find top 10 artists with most Popular Songs
track_by_popularity=Filtered_Track.groupby(['artists','id_artists'],as_index=False)['popularity'].mean()
top_10_Popularity=track_by_popularity.sort_values(by='popularity',ascending=False).head(10)
top_10_Popularity
```

```python
## Artist with Most Popular Songs
Most_Popular=track_by_popularity.sort_values(by='popularity',ascending=False).head(1)
Most_Popular
```

```python
sample_track = track.sample(int(0.005*len(track)))
print(len(sample_track))
```

## Regression Plot Between Loudness and Energy

```python
# now we'll create a regression plot between loudness and energy

plt.figure(figsize=(10,6))
sns.regplot(data = sample_track, y = "loudness", x = "energy", color = "b").set(title = "Loudness vs Energy Correlation")
```

```python
##Analysis
##High Positive correlation between Loudness and Energy  
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_track, y="energy", x="acousticness" , color="c").set(title=" Energy vs Acousticness Correlation")
```

```python
##Analysis
##As the regression line is downwards it indicates that acousticness decreases with increase in energy
```

## Average duration of Songs over the Years

```python
track['duration'] = track['duration_ms'].apply(lambda x: round(x/1000))
track.drop('duration_ms', inplace=True, axis=1)
```

```python
track
```

## BAR PLOT 

```python
track['year'] = track['release_date'].dt.year
track = track[track['year'].notna()]

# Convert 'year' to integer
track['year'] = track['year'].astype(int)

# Create the bar plot without error bars
plt.figure(figsize=(18, 6))  # Set figure size
sns.barplot(x='year', y='duration', data=track, errorbar=None)  # Disable error bars by setting errorbar=None

# Add title and labels
plt.title('Song Duration by Year')
plt.xlabel('Year')
plt.ylabel('Duration (seconds)')

# Rotate the x-axis labels to be vertical
plt.xticks(rotation=90)

# Show plot
plt.show()
```
